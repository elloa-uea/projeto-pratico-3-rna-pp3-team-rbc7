```python
def plotBarChart(data, title):
    fig = plt.subplots(2,2, figsize=(10,20))
    a[0,0] = 
    data = pd.Series(
        data,
        name=title
    )
    freq = data.value_counts().sort_index()
    freq_frame = freq.to_frame()
    mean = data.mean()
    median = data.median()
    #plt.figure(figsize=(12,6))
    plt.bar(freq_frame.index,freq_frame[title],color='#12294B',
            width=6,align='center')

    plt.xlabel('Amostras')
    plt.ylabel('Quantidade')
    plt.title(title)

    plt.axvline(mean, color='r', linestyle='-', label=f"Média: {round(mean,4)}")
    plt.axvline(median, color='orange', linestyle='-', label=f"Mediana: {round(median, 4)}")
    plt.legend()
    #plt.show()
```
