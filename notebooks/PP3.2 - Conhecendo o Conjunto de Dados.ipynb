```python
def plotBarChart(data, title, i):
    ax = plt.subplot(5,2,i)
    data = pd.Series(data, name=title)
    freq = data.value_counts().sort_index()
    freq_frame = freq.to_frame()
    mean = data.mean()
    median = data.median()
    plt.bar(freq_frame.index,freq_frame[title],color='#12294B', width=6,align='center')
    plt.xlabel('Amostras')
    plt.ylabel('Quantidade')
    plt.title(title)

    plt.axvline(mean, color='r', linestyle='-', label=f"Média: {round(mean,4)}")
    plt.axvline(median, color='orange', linestyle='-', label=f"Mediana: {round(median, 4)}")
    plt.legend()
```

```python
fig = plt.figure(figsize=(28,28))
dfList = [df['Elevation'], df['Aspect'], df['Slope'], 
          df['Horizontal_Distance_To_Hydrology'], df['Vertical_Distance_To_Hydrology'], 
          df['Horizontal_Distance_To_Roadways'], df['Hillshade_9am'], df['Hillshade_Noon'], 
          df['Hillshade_3pm'], df['Horizontal_Distance_To_Fire_Points']] 

titleList = ['Elevation', 'Aspect', 'Declive', 'Horizontal Distance To Hydrology', 'Vertical Distance To Hydrology', 
             'Horizontal Distance To Roadways', 'Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm', 'Horizontal_Distance_To_Fire_Points']



for i in range(len(dfList)):
    plotBarChart(dfList[i],titleList[i], i+1)
plt.show()
```

```python
cover_types = df['Cover_Type'].value_counts().sort_index().to_frame()
cover_types['Nome'] = ['Spruce/fir','Lodgepole pine','Ponderosa pine','Cottonwood willow','Aspen','Douglas-fir','Krummholz']
cover_types.set_index('Nome')
ax = cover_types.plot.bar(x='Nome', y='Cover_Type',figsize=(15,6), color='#12294B',  rot=0)
```

A partir da observação do balanceamento do atributo alvo, conseguimos identificar um desbalanceamento entre os valores de saída, com uma concentração de valores maior entre as classificações *Spruce/fir* e *Lodgepole pine*, com aproximadamente 84,8486% das amostras.
